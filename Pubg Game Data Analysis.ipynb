# Solos,Duos and Squad

```python
solos = data[data['numGroups']>50]
duos  = data[(data['numGroups']>25) & (data['numGroups']<=50)]
squad = data[data['numGroups']<=25]
print("Total number of  play solo mode:",len(solos),"Percentage of players:",len(solos)/len(data) * 100)
print("Total number of play duos mode:",len(duos),"Percentage of players:",len(duos)/len(data) * 100)
print("Total number of play squard mode:",len(squad),"Percentage of players:",len(squad)/len(data) * 100)
```

```python
f,ax1 = plt.subplots(figsize=(20,10))
sns.pointplot(x='kills',y='winPlacePerc',data=solos,color='red',alpha=0.8)
sns.pointplot(x='kills',y='winPlacePerc',data=duos,color='blue',alpha=0.8)
sns.pointplot(x='kills',y='winPlacePerc',data=squad,color='green',alpha=0.8)
plt.text(37,0.6,'solos',color='red',fontsize = 17)
plt.text(37,0.55,'duos',color='blue',fontsize = 17)
plt.text(37,0.5,'squad',color='green',fontsize = 17)
plt.xlabel("Number of kills")
plt.ylabel("Win Percentage")
plt.show()
```

```python
f,ax1 = plt.subplots(figsize=(20,10))
sns.pointplot(x='DBNOs',y='winPlacePerc',data=duos,color='red',alpha=0.8)
sns.pointplot(x='DBNOs',y='winPlacePerc',data=squad,color='blue',alpha=0.8)
sns.pointplot(x='assists',y='winPlacePerc',data=duos,color='green',alpha=0.8)
sns.pointplot(x='assists',y='winPlacePerc',data=squad,color='yellow',alpha=0.8)
sns.pointplot(x='revives',y='winPlacePerc',data=duos,color='black',alpha=0.8)
sns.pointplot(x='revives',y='winPlacePerc',data=squad,color='orange',alpha=0.8)
plt.text(14,0.5,'Duos - Assists',color='red')
plt.text(14,0.45,'Duos - DBNOs',color='blue')
plt.text(14,0.4,'Duos - Revives',color='green')
plt.text(14,0.35,'Squads - Assists',color='yellow')
plt.text(14,0.3,'Squads - DBNOs',color='black')
plt.text(14,0.25,'Squads - Revives',color='orange')
plt.xlabel('Duo vs Squad DBNOs, Assists, and Revives')
plt.ylabel('WinPlacePerc')
plt.title('Duo vs Squad DBNOs, Assists, and Revives')
plt.grid()
plt.show()
```
